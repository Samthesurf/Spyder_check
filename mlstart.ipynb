```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor

# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
life_sat = pd.read_csv(data_root+'lifesat/lifesat.csv')
life_sat.columns = life_sat.columns.str.replace(' ,', '')
X = life_sat[["GDP per capita (USD)"]].values
Y = life_sat[["Life satisfaction"]].values
# Data visualization
life_sat.plot(kind='scatter', grid =True, x ="GDP per capita (USD)", y ="Life satisfaction")
plt.axis([23_500,62_500,4,9])
plt.show()
# select a linear model
model = KNeighborsRegressor(n_neighbors=3)
# training time
model.fit(X, Y)
# make a new prediction
X_new = [[37_655.2]]
print(model.predict(X_new))
```

```python
import numpy as np
import matplotlib.pyplot as plt
# cumsum() creates an array
one_dim_array = np.cumsum(np.random.randn(17,1))
plt.plot(one_dim_array,marker='X',markersize=12,markerfacecolor='#2596be',color='#eab676',label="Line chart")
plt.grid(True)
plt.title('Simple line-chart with Markers')
plt.show()
```
